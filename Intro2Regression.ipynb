```python
X_train = X[:40] # First 40 samples
X_test = X[40:] # Last 10 samples

y_train = y[:40]
y_test = y[40:]
```

```python
plt.scatter(X_train,y_train,color = 'red', label = 'training data')
plt.scatter(X_test,y_test,color = 'blue', label = 'testing data')
plt.legend()
```

```python
# 1. Create the model - input, output and hidden layer
tf.random.set_seed(12)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation = None, name = 'hidden_layer_1'),
    tf.keras.layers.Dense(1, name = 'output_layer'),
])

# 2. Compile the model - define loss and optimizer
model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
    metrics = ['mae'])

# 3. Fit the model
model.fit(tf.expand_dims(X,1),y, epochs = 100)
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model = model, show_shapes= True)
```

```python
y_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
def plot_predictions(train_data = X_train, train_labels = y_train, test_data = X_test, test_labels = y_test, predictions = y_pred):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data,train_labels, c ='b', label = 'training data')
    plt.scatter(test_data,test_labels, c ='r', label = 'testing data')
    
    plt.scatter(test_data,predictions, c = 'g', label = 'predictions')
    
    plt.legend()
    
    
plot_predictions()
```

```python
model.evaluate(X_test, y_test)
```

```python
mae = tf.metrics.mean_absolute_error(
    y_true = y_test,
    y_pred = tf.squeeze(y_pred))
mae
```
